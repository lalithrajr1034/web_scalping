```python
soup = BeautifulSoup(lalith.content, "html.parser")
```

```python
soup.prettify()
```

```python
la = soup.find_all("a")
for i in la:
    tit = i.get("title")
    if tit:
        # clean_title = tit.split(":")[0].split(",")[0].strip()
        print(tit)
```

```python

la = soup.find_all("a")
for i in la:
    tit = i.get("title")
    if tit:
        clean_title = tit.split(":")[0].split(",")[0].strip()
        print(clean_title)
```
